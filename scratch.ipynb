```python
from PIL import Image
from typing import List, Tuple


def random_mouse_movement(desktop, num_moves=5) -> List[Tuple[Image.Image, int, int]]:
    import random
    import time

    out = []
    # Screen dimensions from your screenshot
    width, height = 1024, 768

    for _ in range(num_moves):
        # Generate random coordinates within screen bounds
        x = random.randint(0, width)
        y = random.randint(0, height)

        # Move the mouse to the random position
        desktop.move_mouse(x, y)

        shots = desktop.take_screenshots()
        out.append((shots[-1], x, y))

    return out
```

```python
shots = random_mouse_movement(desktop, 5)
```
